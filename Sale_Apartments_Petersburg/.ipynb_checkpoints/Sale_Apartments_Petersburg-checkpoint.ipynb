Для заполнения некоторых пропусков лучше брать лучше брать данные по группам, будем ориентироваться на локацию, для заполнения **высоты потолков** и **этажность**. Но в начале проверим на дубликаты в названиях населенных пунктов.

Видим занчение Nan посчитаем их кол-во

```python
print('Объявлений с пропуском населенного пункта:',df['locality_name'].isna().sum())
print('Процент объявлений с пропуском населенного пункта:', round(df['locality_name'].isna().sum()/df.shape[0]*100,2),'%')
```

Удалим 49 строк так как это не сильно скажется на общих данных так как это всего 0,2%

```python
df = df.dropna(subset=['locality_name'])
```

Напишем функцию которая отбросит все до заглавной буквы и провереим на тестовой серии:

```python
def crop_name_city(row):
    
    text = row['locality_name']
    
    try:
        for x in range(len(text)):
            if  'Я' >=  text[x] != ' ':
                return text[x:]
            
    except:
        display('Ошибка проверить данные в `locality_name`')

#для проверки сделаем серию pandas и запустим нашу функцию
row_values = ['деревня Нижние Осельк']
row_columns = ['locality_name']

row = pd.Series(data=row_values, index=row_columns)

crop_name_city(row)
```

Применим функцию на нашу базу и посчитаем и выведем уникальные имена.

```python
df['locality_name'] = df.apply(crop_name_city, axis=1)
print('Городов осталось:',len(df['locality_name'].sort_values().unique()))
df['locality_name'].sort_values().unique()
```

Заполним *медиальным значением* ячейки по **высоте потолков**, для количественных данных это оптимальное решение, группируя данные удаленности от центра, затем по населенным пунктам.

После посчитаем количество пустый ячеек в колонке:

```python
df['city_center_km'] = round(df['city_сenters_nearest'] / 1000)# добавим колонку чтобы срупировать по удаленности
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('city_center_km')['ceiling_height'].transform("median"))

df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('locality_name')['ceiling_height'].transform("median")) #замена по группам
df.loc[df['ceiling_height'] .isna(),'ceiling_height'] = df['ceiling_height'].median() #замена оставшихся медиал по всей таблице
df['ceiling_height'].isna().sum()
```

Заполним *медиальным значением* "этажность" здания, стоит учесть что при заполнение "этажности" нельзя допустить чтобы этажность здания была меньше чем этаж квартиры в самом объявлении.

Далее посчитаем количество пустый ячеек в колонке:

```python
df['floors_total'] = df['floors_total'].fillna(df.groupby('city_center_km')['floors_total'].transform("median"))

df['floors_total'] = df['floors_total'].fillna(df.groupby('locality_name')['floors_total'].transform("median"))#замена по группам
#df.loc[df['floors_total'].isna(),'floors_total'] = df['floors_total'].median()#замена оставшихся медиал по всей таблице
df.loc[df['floors_total'] < df['floor'],'floors_total'] = df['floor']#если всего этажей меньше этажа в объявлении меняем общее колчество этажей
df['floors_total'].isna().sum()
```

Пропуски в жилой площади и полощади кухни заполним медиальным значением взяв значения в зависимоти от количества комнат, это будет оптимальным вариантом. Посмотрим сколько всего ячеек пустых:

Напишем функцию `floor_group` для классификации по этажу:

- первый
- поледний
- другой

Проверим ее на тестовом датафрейме:

```python
def floor_group(row):
    
    floor = row['floor']
    floors_total = row['floors_total']
    
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'


#для проверки сделаем серию pandas и запустим нашу функцию
row_values = [1,5]
row_columns = ['floor','floors_total']

row = pd.Series(data=row_values, index=row_columns)

floor_group(row)
```

Применим нашу функцию к нашей базе и выведем первые 5 ячеек для проверки:

```python
df['floor_category'] = df.apply(floor_group, axis=1)
df.head(5)
```

Добавим соотношение жилой и общей площади, а также отношение площади кухни к общей и посмотрим первые 5 ячеек для проверки:

```python
df['ratio_living_total'] = round(df['living_area'] / df['total_area'] * 100)
df['ratio_kitchen_total'] = round(df['kitchen_area'] / df['total_area'] * 100)
df.head()
```

Посмотрим влияние дня недели на стоимость и на продолжительность продажи:

```python
crop_df['weekday'] = crop_df['first_day_exposition'].dt.weekday

(
    crop_df
    .groupby(['weekday'])
    .agg({'price_m2':'median'}).
    plot( y='price_m2', style='o-',  grid=True, figsize=(14, 3))
)

plt.title('График стоимости кв.м в зависимости от размещения объявления в определенный день недели', size=14)
plt.xticks(rotation=0)
plt.xlabel('День недели', size=12)
plt.ylabel('Стоимость кв.м.', size=12)
plt.show()

(
    crop_df
    .groupby(['weekday'])
    .agg({'days_exposition':'median'}).
    plot( y='days_exposition', style='o-',  grid=True, figsize=(14, 3), color='#168337')
)

plt.title('График зависимости продолжительности продажи от размещения в определенный день недели', size=14)
plt.xticks(rotation=0)
plt.xlabel('День недели', size=12)
plt.ylabel('Кол-во дней', size=12)
plt.show()
```

На первом графике видим небольшое увеличение с 94тр. недели вторник и среды до 95тр. и постепенное снижение к концу недели до 93тр. в субботу.

На втором графике что в среднем срок размещения объявления 91 день с небольшим увеличением до 94 дней в среду.

Видимо просмотром объявлений занимаются в начале недели, либо в конце неделе в середине интерес к новым объявлениям снижен.
В тоже время средняя цена выше именно при выкладывание объявления в середине недели.

Посмотрим влияние месяца на стоимость и на продолжительность продажи:

```python
crop_df['month'] = crop_df['first_day_exposition'].dt.month

(
    crop_df
    .groupby(['month'])
    .agg({'price_m2':'median'}).
    plot( 
        y='price_m2', 
        style='o-',  
        grid=True, 
        figsize=(12, 3)
    )
)
plt.title('График стоимости кв.м в зависимости от размещения объявления в определенный месяц', size=14)
plt.xticks(rotation=0)
plt.xlabel('Месяц', size=12)
plt.ylabel('Стоимость кв.м.', size=12)
plt.show()


(
    crop_df
    .groupby(['month'])
    .agg({'days_exposition':'median'}).
    plot( 
        y='days_exposition', 
        style='o-',
        grid=True,
        figsize=(12, 3), 
        color='#168337'        
    )   
)

plt.title('График зависимости продолжительности продажи от размещения в определенный месяц', size=14)
plt.xticks(rotation=0)
plt.xlabel('Месяц', size=12)
plt.ylabel('Кол-во дней', size=12)
plt.show()
```

Цены растут с начала года до апреля, затем май и июнь средняя цена падает к минимуму года (видимо летние отпуска влияют на рынок недвижимости) к августу и сентябрю цена опять растет, затем снижается к концу года.

Так же продолжительность продажи увеличена как раз там где провал в средней стоимости кв.м. тоесть в май и июнь не самое удачное время для продажи квартиры и цены снижают чтобы продать в этот период.

Посмотрим влияние года на стоимость и на продолжительность продажи:

```python
crop_df['year'] = crop_df['first_day_exposition'].dt.year

(
    crop_df
    .groupby(['year'])
    .agg({'price_m2':'median'}).
    plot( 
        y='price_m2', 
        style='o-',  
        grid=True, 
        figsize=(12, 3)
    )
)

plt.title('График зависимости продолжительности продажи от размещения в определенный год', size=14)
plt.xticks(rotation=0)
plt.xlabel('Год', size=12)
plt.ylabel('Стоимость кв.м.', size=12)
plt.show()

(
    crop_df
    .groupby(['year'])
    .agg({'days_exposition':'median'}).
    plot( 
        y='days_exposition', 
        style='o-',
        grid=True,
        figsize=(12, 3),
        color='#168337'  
    )   
)

plt.title('График зависимости продолжительности продажи от размещения в определенный год', size=14)
plt.xticks(rotation=0)
plt.xlabel('Год', size=12)
plt.ylabel('Кол-во дней', size=12)
plt.show()
```

Теперь выведем общую информацию по средней стоимости из двух баз и посмотрим разницу:

```python
list_col = ['price_m2','floor','rooms','city_center_km','ceiling_height','total_area']

for col in list_col:
    crop_df_describe = round(crop_df[col].describe())
    center_piter_df_describe = round(center_piter_df[col].describe())

    summary_describe = pd.concat([crop_df_describe, center_piter_df_describe], axis=1)# соеденим две серии в один датафрейм

    summary_describe.columns = ['common_data','center_data']# переменуем колонки
    summary_describe['difference'] = round((1 - summary_describe['common_data'] / summary_describe['center_data']) * 100)# посчитаем разницу в %
    display(summary_describe)
    display(col)
```

Если сравнить базы в целом то разница будет в районе 20% стоимость жилья в центре выше чем в общем по Санкт-Петербургу.
